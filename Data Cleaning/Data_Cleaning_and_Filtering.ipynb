### Missing Values

```python
print("\nMissing Values in Event Data:")
print(event_data.isnull().sum())

print("\nMissing Values in Fight Data:")
print(fight_data.isnull().sum())

print("\nMissing Values in Fight Stat Data:")
print(fight_stat_data.isnull().sum())

print("\nMissing Values in Fighter Data:")
print(fighter_data.isnull().sum())
```

event_data: "Unknown" since this isn't critical information in my opinion

```python
event_data['event_state'].fillna('Unknown', inplace=True)
```

referee, weight_class, result_details: "Not Available"

f_1, f_2, winner: Fill with 0 or drop rows? Probably drop

```python
fight_data[['referee', 'weight_class', 'result_details']] = fight_data[
    ['referee', 'weight_class', 'result_details']].fillna('Not Available')
fight_data[['f_1', 'f_2', 'winner']] = fight_data[['f_1', 'f_2', 'winner']].fillna(0)
```

fighter_id: Fill with 0 because we probably do not need this, but we can drop if needed. 
fight stats (knockdowns, total_strikes_att): Fill with 0

```python
fight_stat_data['fighter_id'].fillna(0, inplace=True)
fight_stat_data.fillna(0, inplace=True)
```

fighter_l_name: Unknown

fighter_nickname, fighter_stance: Not Available

fighter_height_cm, fighter_weight_lbs, fighter_reach_cm, fighter_dob: maybe median value? probably drop 

fighter_nc_dq: Fill with 0

```python
fighter_data['fighter_l_name'].fillna('Unknown', inplace=True)
fighter_data[['fighter_nickname', 'fighter_stance']] = fighter_data[
    ['fighter_nickname', 'fighter_stance']].fillna('Not Available')

fighter_data['fighter_height_cm'].fillna(fighter_data['fighter_height_cm'].median(), inplace=True)
fighter_data['fighter_weight_lbs'].fillna(fighter_data['fighter_weight_lbs'].median(), inplace=True)
fighter_data['fighter_reach_cm'].fillna(fighter_data['fighter_reach_cm'].median(), inplace=True)

fighter_data['fighter_dob'].fillna('1900-01-01', inplace=True)

fighter_data['fighter_nc_dq'].fillna(0, inplace=True)
```

### Converting Data Type
event_date in Event Data = datetime64[ns]

fighter_dob in Fighter Data = datetime64[ns]

date and dob in Master Dataframe = datetime64[ns]

f_1, f_2, winner in Fight Data = float64

knockdowns, total_strikes_att, submission_att in Fight Stat Data = float64

```python
event_data['event_date'] = pd.to_datetime(event_data['event_date'], errors='coerce')
fighter_data['fighter_dob'] = pd.to_datetime(fighter_data['fighter_dob'], errors='coerce')
master_dataframe['date'] = pd.to_datetime(master_dataframe['date'], errors='coerce')
master_dataframe['dob'] = pd.to_datetime(master_dataframe['dob'], errors='coerce')


fight_data['f_1'] = pd.to_numeric(fight_data['f_1'], errors='coerce')
fight_data['f_2'] = pd.to_numeric(fight_data['f_2'], errors='coerce')
fight_data['winner'] = pd.to_numeric(fight_data['winner'], errors='coerce')
fight_stat_data[['fighter_id', 'knockdowns', 'total_strikes_att', 
                 'total_strikes_succ', 'sig_strikes_att', 'sig_strikes_succ', 
                 'takedown_att', 'takedown_succ', 'submission_att', 'reversals']] = \
fight_stat_data[['fighter_id', 'knockdowns', 'total_strikes_att', 
                 'total_strikes_succ', 'sig_strikes_att', 'sig_strikes_succ', 
                 'takedown_att', 'takedown_succ', 'submission_att', 'reversals']].apply(pd.to_numeric, errors='coerce')


event_data_dtypes = event_data.dtypes
fighter_data_dtypes = fighter_data.dtypes
fight_data_dtypes = fight_data.dtypes
fight_stat_data_dtypes = fight_stat_data.dtypes
master_dataframe_dtypes = master_dataframe.dtypes

event_data_dtypes, fighter_data_dtypes, fight_data_dtypes, fight_stat_data_dtypes, master_dataframe_dtypes
```

### Inconsistent Data Entry
"orthadox" to "orthodox"

open stance = open_stance

women's weight class = w- 

```python

fighter_data['fighter_stance'] = fighter_data['fighter_stance'].str.lower().replace(
    {'orthadox': 'orthodox', 'open stance': 'open_stance', 'sideways': 'sideways'})

fight_data['weight_class'] = fight_data['weight_class'].str.lower().replace(
    {"women's strawweight": "w-strawweight", 
     "women's bantamweight": "w-bantamweight",
     "women's flyweight": "w-flyweight",
     "women's featherweight": "w-featherweight",
     'catch weight': 'catch_weight', 
     'open weight': 'open_weight'})


master_dataframe['stance'] = master_dataframe['stance'].str.lower().replace(
    {'orthadox': 'orthodox', 'open stance': 'open_stance', 'sideways': 'sideways'})


fighter_stance_unique_cleaned = fighter_data['fighter_stance'].unique()
weight_class_unique_cleaned = fight_data['weight_class'].unique()
master_stance_unique_cleaned = master_dataframe['stance'].unique()

fighter_stance_unique_cleaned, weight_class_unique_cleaned, master_stance_unique_cleaned
```

### Merge Datasets:

Again, need to discuss. 

example: 

Fight Data + Fight Stat Data: Merged on fight_id using inner join.

Merged with Fighter Data: Linked using f_1 as the fighter_id from Fighter Data (main fighter for the fight).

Used left join to ensure no missing fighters from the fight data.


```python

fighter_data['fighter_f_name'] = fighter_data['fighter_f_name'].str.strip().str.lower()
fighter_data['fighter_l_name'] = fighter_data['fighter_l_name'].str.strip().str.lower()

fight_data_cleaned['fight_id'] = fight_data_cleaned['fight_id'].astype(int)
fight_stat_data_cleaned['fight_id'] = fight_stat_data_cleaned['fight_id'].astype(int)
fighter_data_cleaned['fighter_id'] = fighter_data_cleaned['fighter_id'].astype(int)

merged_fight_data = pd.merge(fight_data_cleaned, fight_stat_data_cleaned, on='fight_id', how='inner')

merged_fight_fighter_data = pd.merge(merged_fight_data, fighter_data_cleaned, left_on='f_1', right_on='fighter_id', how='left')


merged_shape = merged_fight_fighter_data.shape
merged_preview = merged_fight_fighter_data.head()

merged_shape, merged_preview
```
